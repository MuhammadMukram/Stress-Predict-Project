```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle
from sklearn import metrics
```

```python
data = "SaYoPillow.csv"
dataset = pd.read_csv(data)
```

```python
dataset.head()
```

```python
dataset = dataset.rename(columns={'sr.1' : 'sh'})
dataset.head()
```

```python
dataset.info()
```

```python
dataset['sl'].unique()
```

```python
dataset.sl.value_counts()
```

```python
qualityType = { 0 : 0, 1 : 0, 2 : 1, 3 : 2, 4 : 2 }
dataset["sl"] =  dataset["sl"].map(qualityType)
dataset.sl.value_counts()
```

<h3>Note :</h3>
<p>
Transfrom stress Level to 3 types: 
<ul> 
<li> Not Stressed : 0 </li>
<li> Stressed : 1 </li>
<li> Very Stressed : 2 </li>
</ul>
</p>

```python
fig = px.box(dataset)

fig.show()
```

<p>
Karena tidak terdapat data outlier, maka dapat langsung dibagi antara data train dan data test nya.
</p>

```python
X = dataset.drop('sl',axis=1).values
y = dataset['sl'].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.25, random_state=4)
```

<h1> DECISION TREE </h1>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
stressTree = DecisionTreeClassifier(criterion="gini", max_depth = 10)
stressTree.fit(X_trainset, y_trainset)
```

```python
predTree = stressTree.predict(X_testset)
```

```python
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree)*100, '%')
```

```python
# Export Decision Tree Model
filename = 'dtree_pickle'
pickle.dump(stressTree, open(filename, 'wb'))
```

<h3>
# DECISION TREE EVALUATION
</h3>

```python
## R2
from sklearn.metrics import r2_score
r2 = r2_score(y_testset, predTree) * 100
print("R-Square : ", r2, '%')
```

```python
## MSE
from sklearn.metrics import mean_squared_error
print("MSE : ",mean_squared_error(y_testset, predTree))
```

<h1> SVM </h1>

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel='linear', gamma='auto').fit(X_trainset,y_trainset) 
```

```python
y_predict_scm = clf.predict(X_testset)
```

```python
print("SVM's Accuracy: ", round(metrics.accuracy_score(y_testset, y_predict_scm)*100,2),"%")
```

```python
# Export SVM Model
filename = 'svm_pickle'
pickle.dump(clf, open(filename, 'wb'))
```

<h3>
# SVM EVALUATION
</h3>

```python
## R2
from sklearn.metrics import r2_score
r2 = r2_score(y_testset, y_predict_scm) * 100
print("R-Square : ", r2, "%")
```

```python
## MSE
from sklearn.metrics import mean_squared_error
print("MSE : ",mean_squared_error(y_testset, y_predict_scm))
```
